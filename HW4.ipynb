#Question 3

1. By repeatedly resampling from the original dataset and computing the mean for each resample, create the bootstrapped sample means.
2. To determine the 2.5th and 97.5th percentiles of the bootstrapped means, use the np.quantile(...) method. This represents the central 95% of the sample mean distribution:
Lower bound = np.quantile(bootstrapped_means, 0.025)
Upper bound = np.quantile(bootstrapped_means, 0.975)

```python
#Question 4

import numpy as np

# Example sample data
sample = np.array([5, 7, 9, 12, 15, 17, 19])

# Number of bootstrap samples
n_iterations = 10000

# Empty list to store the bootstrapped means
bootstrapped_means = []

# Bootstrapping process for the mean
for _ in range(n_iterations):
    # Generate a bootstrap sample by resampling with replacement
    bootstrap_sample = np.random.choice(sample, size=len(sample), replace=True)
    # Calculate the mean of the bootstrap sample and store it
    bootstrapped_means.append(np.mean(bootstrap_sample))

# Convert list of bootstrapped means to a NumPy array for quantile calculation
bootstrapped_means = np.array(bootstrapped_means)

# Calculate the 95% confidence interval for the mean (2.5th and 97.5th percentiles)
mean_conf_interval = np.quantile(bootstrapped_means, [0.025, 0.975])

# Print the 95% confidence interval for the mean
print(f"95% Bootstrap Confidence Interval for the Mean: {mean_conf_interval}")

# To produce a 95% bootstrap confidence interval for the population median, 
# simply replace the np.mean() in the bootstrapping process with np.median():

# Empty list to store the bootstrapped medians
bootstrapped_medians = []

# Bootstrapping process for the median
for _ in range(n_iterations):
    # Generate a bootstrap sample by resampling with replacement
    bootstrap_sample = np.random.choice(sample, size=len(sample), replace=True)
    # Calculate the median of the bootstrap sample and store it
    bootstrapped_medians.append(np.median(bootstrap_sample))

# Convert list of bootstrapped medians to a NumPy array for quantile calculation
bootstrapped_medians = np.array(bootstrapped_medians)

# Calculate the 95% confidence interval for the median (2.5th and 97.5th percentiles)
median_conf_interval = np.quantile(bootstrapped_medians, [0.025, 0.975])

# Print the 95% confidence interval for the median
print(f"95% Bootstrap Confidence Interval for the Median: {median_conf_interval}")
```

#Question 5

Confidence intervals require that one distinguish between a sample statistic and a population parameter since they have different functions. The sample statistic is a computed value from a sample (such as a sample mean), which acts as an estimate of the population parameter. The population parameter is the fixed, unknown value we are seeking to estimate. In order to give a range of values likely to contain the population parameter while taking sampling variability into account, confidence intervals are constructed around the sample statistic.

#Question 8

1. We want to assess the efficacy of an AliTech-developed vaccine in this analysis. Finding out if the vaccination improves the final health score over the starting score is something we are really interested in. In this case, the null hypothesis (H0) is that the vaccine has no effect on the health scores. This indicates that, on average, there should be no variation between the initial and final health scores. 

2. After that, the null hypothesis (H0) posits that the vaccine has no impact on the health of patients, i.e., the final health scores are not significantly different from the initial health scores. This hypothesis will be tested against the alternative hypothesis (H1), which suggests that the vaccine does have an effect, leading to a change in health scores.

3. To begin with, see the distribution of the baseline and end health ratings visually to gain some knowledge.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reading the CSV file
data = pd.read_csv('vaccine_data.csv')

# Visualizing the Initial and Final Health Scores
plt.figure(figsize=(10, 6))
sns.kdeplot(data['InitialHealthScore'], label='Initial Health Score', shade=True)
sns.kdeplot(data['FinalHealthScore'], label='Final Health Score', shade=True)
plt.title('Distribution of Initial and Final Health Scores')
plt.xlabel('Health Score')
plt.ylabel('Density')
plt.legend()
plt.show()
```

This plot will help us visually inspect whether there is a noticeable difference between the initial and final health scores.

4. Quantitative Analysis: To determine the distribution of the variation between the starting and end health ratings, we will employ bootstrapping. We can determine if the vaccine had a statistically significant effect by resampling with replacement, computing the mean difference for each sample, and estimating a confidence interval.

5. In order to determine the confidence interval of the mean difference in scores, let's apply bootstrapping on the data.

```python
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42)

# Calculate the difference between final and initial health scores
data['HealthScoreDifference'] = data['FinalHealthScore'] - data['InitialHealthScore']

# Bootstrapping
n_bootstraps = 10000
boot_means = []

for _ in range(n_bootstraps):
    # Resample with replacement
    bootstrap_sample = data.sample(n=len(data), replace=True)
    # Calculate the mean difference for this bootstrap sample
    mean_difference = bootstrap_sample['HealthScoreDifference'].mean()
    boot_means.append(mean_difference)

# Convert the list to a numpy array for analysis
boot_means = np.array(boot_means)

# Calculate the 95% confidence interval
ci_lower = np.quantile(boot_means, 0.025)
ci_upper = np.quantile(boot_means, 0.975)

print(f"95% Confidence Interval for the mean difference: [{ci_lower}, {ci_upper}]")
```

6. Visualizing the distribution of the bootstrapped mean differences.

```python
plt.figure(figsize=(10, 6))
sns.histplot(boot_means, bins=30, kde=True)
plt.axvline(ci_lower, color='red', linestyle='--', label=f'Lower CI: {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='--', label=f'Upper CI: {ci_upper:.2f}')
plt.title('Bootstrapped Distribution of Health Score Differences')
plt.xlabel('Mean Difference in Health Scores')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
